```
import pandas as pd

url = "https://raw.githubusercontent.com/lvlanson/uebersterblichkeit/master/BL_2016_2020.csv"



df = pd.read_csv(url, sep=",", skiprows=8)

columns = list(df.columns)

columns[1] = "Jahr"
columns[2] = "Bundesland"

df.columns = columns

df_sachsen = df.loc[df["Bundesland"] == "Sachsen "]

df_sachsen.head()
```

```
from plotly.subplots import make_subplots
import plotly.graph_objects as go

data = []
years = []

for row in range(df_sachsen.shape[0]):
  trace = go.Scatter(x=df_sachsen.columns[3:-1].values, 
                     y=df_sachsen.iloc[row:row+1,3:-1].values[0],
                     name=str(df_sachsen.Jahr.values[row]))
  if df_sachsen.Jahr.values[row] != 2020:
    years.append(df_sachsen.iloc[row:row+1,3:-1].values[0])

  data.append(trace)
fig = go.Figure(data=data)
fig.show()
```

```
import copy

mit = []
for year in years:
  mittel = 0
  counter = 0
  for day in year:
    if isinstance(day, int):
      mittel += day
      counter += 1
  mittel /= counter
  mit.append(mittel)


years_fix = copy.deepcopy(years)
for i,year in enumerate(years_fix):
  for j,day in enumerate(year):
    if isinstance(day, str):
      years_fix[i][j] = mit[i]

y_mit = []

for y_1, y_2, y_3, y_4 in zip(*years_fix):
  y_mit.append((y_1+y_2+y_3+y_4)/4)
```

```
from plotly.subplots import make_subplots
import plotly.graph_objects as go

data = []
years = []

row = 0

trace = go.Scatter(x=df_sachsen.columns[3:-1].values, 
                   y=df_sachsen.iloc[row:row+1,3:-1].values[0],
                   name=str(df_sachsen.Jahr.values[row]))
data.append(trace)

trace = go.Scatter(x=df_sachsen.columns[3:-1].values, 
                   y=y_mit,
                   name="Durchschnitt 2016-2019")
data.append(trace)
fig = go.Figure(data=data)
fig.show()
```
